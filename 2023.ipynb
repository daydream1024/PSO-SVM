```python
import pandas
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
import tkinter
from sklearn import tree
import pyswarms as ps
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)#控制变量
```

```python
names = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']
HeartDisease = pandas.read_csv('./processed.cleveland.csv',header=None,names=names)
```

```python
median=HeartDisease['ca'].median()
HeartDisease['ca'].fillna(median,inplace=True)
median=HeartDisease['thal'].median()
HeartDisease['thal'].fillna(median,inplace=True)
#HeartDisease.info()
```

```python
label = HeartDisease['num']
data = HeartDisease.drop(['num'],axis=1)
```

```python
if data.shape[0] == label.shape[0]:
    print('The two result files have written to the Output folder! ')
else:
    print('Sample number not equal to label number')
    exit(-1)
```

```python
clf = tree.DecisionTreeClassifier()
acc  = cross_val_score(clf, data,label, cv=10)
print(np.mean(acc))
```

```python
clf=SVC()
acc  = cross_val_score(clf, data,label, cv=10)
print(np.mean(acc))
```

```python
x_train , x_test , y_train , y_test =train_test_split(data, label, test_size=0.25, train_size=0.75, random_state=42)
```

```python
def plot(position):
    x = []
    y = []
    for i in range(0,len(position)):
        x.append(position[i][0])
        y.append(position[i][1])
    colors = (0,0,0)
    plt.scatter(x, y, c = colors, alpha = 0.1)
    plt.xlabel('C')
    plt.ylabel('gamma')
    plt.axis([0,1,0,1])
    plt.gca().set_aspect('equal', adjustable='box')
    return plt.show()
```

```python
#求取最优解的方程。
def func(params,X,y):
    scores = np.zeros((10, 3))
    for i in range(10):
        clf = SVC
        scores[i, :] = cross_val_score(clf(C=params[i][0], gamma=params[i][1]),X, y, cv=3)
    plot(params)
    return 1-np.mean(scores, axis=1)
# Set optimization parameters
bounds = ([0, 0], [1, 1])# lower bounds of C and gamma and upper bounds of C and gamma
options = {'c1': 2, 'c2': 2, 'w': 0.7}#c1为粒子的个体认知系数，c2为群体系数系数，w为惯性因子。

#n_particles:粒子群中的粒子数。
#dimensions：待求取参数的维度。
#bounds:待求参数的取值范围。
optimizer = ps.single.GlobalBestPSO(n_particles=10, dimensions=2, options=options,bounds=bounds)

# Run optimization
best_cost, best_params = optimizer.optimize(func, iters=500, X=data, y=label)

# Print results

print('Best cost:', best_cost)#最小代价
print('Best parameters:', best_params)#最好位置
```

```python
clf=SVC(C=best_params[0],gamma=best_params[1])
psoacc  = cross_val_score(clf, data,label, cv=3).mean()
print(psoacc)
```
